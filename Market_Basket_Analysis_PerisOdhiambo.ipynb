#Data Preprocessing

```python
from google.colab import drive
import pandas as pd

drive.mount('/content/drive')

csv_file_path = '/content/drive/MyDrive/west_end.csv'
```

```python
# Load the dataset
data = pd.read_csv(csv_file_path, error_bad_lines=False, header=None)

column_names = [
    'Store', 'Report Type', 'Page Info', 'Start Date Header', 'Start Date',
    'End Date Header', 'End Date', 'Date', 'Till No header', 'Cashier header',
    'Tran No Header', 'Product header', 'Barcode header', 'Sales Header', 'Quantity',
    'Sold', 'Date and Time', 'Till No', 'Cashier', 'Transaction No', 'Product Code',
    'Product Name', 'Barcode', 'Sales VALUE', 'Quantity Sold', 'Report Totals', 'RPT1',
    'RPT2', 'PTAudit', 'Printed by', 'Print Date', 'Print time', 'Nothing'
]

data.columns = column_names
```

```python
pd.set_option('display.max_columns', None)

print(f"Current number of columns in DataFrame: {data.shape[1]}")

data.head(100)
```

```python
# Drop columns that are not needed
new_data = data[['Date and Time', 'Till No', 'Cashier', 'Transaction No', 'Product Code',
                     'Product Name', 'Barcode', 'Sales VALUE', 'Quantity Sold']]

# Display the first few rows to verify the selection
new_data.head(100)
```

```python
#Dropping first row due to misalignment
new_data = new_data.iloc[1:].reset_index(drop=True)

new_data.head(100)
```

```python
#Convert the 'Date and Time' column to datetime format
new_data['Date and Time'] = pd.to_datetime(new_data['Date and Time'], errors='coerce')

#Drop rows with missing values
new_data = new_data.dropna()

#Reset index after dropping rows
new_data.reset_index(drop=True, inplace=True)

#covert "Sales VALUE" to numeric data
new_data['Sales VALUE'] = pd.to_numeric(data['Sales VALUE'], errors='coerce')
```

# Products that sell fastest

```python
# Aggregating sales data by product to determine the total quantity sold for each item
product_sales = new_data.groupby('Product Name')['Quantity Sold'].sum().reset_index()

# Sorting the products based on the total quantity sold in descending order
top_selling_products = product_sales.sort_values(by='Quantity Sold', ascending=False).reset_index(drop=True)

# Display the top 10 selling products
top_selling_products.head(10)
```

# Frequently being bought together

I tried the other library with apriori, mlxtend which I have been practicing as I experienced runtime problems with apyori


```python
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
# Creating a list of transactions, where each transaction is a list of items bought together
transactions = data.groupby('Transaction No')['Product Name'].apply(list).values.tolist()

# Initialize TransactionEncoder
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)

# Applying Apriori algorithm to find frequent itemsets with a minimum support threshold
frequent_itemsets = apriori(df, min_support=0.0001, use_colnames=True)

# Sorting the itemsets by support in descending order
frequent_itemsets = frequent_itemsets.sort_values(by='support', ascending=False).reset_index(drop=True)

pd.set_option('display.max_colwidth', None)

frequent_itemsets.head(10)
```

# Highest lift ratio

```python
from mlxtend.frequent_patterns import association_rules
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

# Filtering the rules where lift is greater than or equal to 3
rules_with_lift_three = rules[rules['lift'] >= 3].sort_values(by='lift', ascending=False).reset_index(drop=True)

rules_with_lift_three['Rule'] = rules_with_lift_three['antecedents'].astype(str) \
                                + " --> " \
                                + rules_with_lift_three['consequents'].astype(str) \


# Define the columns to display, including the new 'Rule' column
columns_to_display = ['Rule', 'support', 'confidence', 'lift']

# Display the formatted rules
print(rules_with_lift_three[columns_to_display])
```

# Highest confidence values

```python
# Filter the rules for those with 100% confidence
rules_with_full_confidence = rules[rules['confidence'] == 1].sort_values(by='lift', ascending=False).reset_index(drop=True)

# Creating the 'Rule' column
rules_with_full_confidence['Rule'] = rules_with_full_confidence['antecedents'].astype(str) \
                                     + " --> " \
                                     + rules_with_full_confidence['consequents'].astype(str)

columns_to_display = ['Rule', 'support', 'confidence', 'lift']

# Displaying the first 10 formatted rules
print(rules_with_full_confidence[columns_to_display].head(10))
```

#Items for the same shelf

```python
high_confidence_rules = rules[rules['confidence'] >= 0.8]

# Function to extract items from a frozenset and return a list
def extract_items(frozenset_item):
    return list(frozenset_item)

# Extract and aggregate items
shelf_recommendations = []

for index, row in high_confidence_rules.iterrows():
    items_together = extract_items(row['antecedents']) + extract_items(row['consequents'])
    shelf_recommendations.append(items_together)

# To get rid of redundancy
unique_recommendations = set(tuple(sorted(items)) for items in shelf_recommendations)

# Print unique shelf placement recommendations
for recommendation in unique_recommendations:
    print(recommendation)
```
